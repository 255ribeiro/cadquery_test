```python
import cadquery as cq
import ipywidgets as widgets
from ocp_vscode import show

def make_window(width, height, depth, thickness):
    outer = cq.Workplane("XY").rect(width, height).extrude(depth)
    inner = cq.Workplane("XY").rect(width - thickness*2, height - thickness*2).extrude(depth)
    return outer.cut(inner)




w = widgets.FloatSlider(value=200, min=50, max=400, step=10, description="Width")
h = widgets.FloatSlider(value=150, min=50, max=400, step=10, description="Height")
d = widgets.FloatSlider(value=20,  min=5,  max=100, step=5,  description="Depth")
t = widgets.FloatSlider(value=20,  min=5,  max=50,  step=1,  description="Thickness")

def update(change):
    shape = make_window(w.value, h.value, d.value, t.value)
    show(shape)

w.observe(update, names="value")
h.observe(update, names="value")
d.observe(update, names="value")
t.observe(update, names="value")

display(widgets.VBox([w, h, d, t]))
show(make_window(w.value, h.value, d.value, t.value))
```
